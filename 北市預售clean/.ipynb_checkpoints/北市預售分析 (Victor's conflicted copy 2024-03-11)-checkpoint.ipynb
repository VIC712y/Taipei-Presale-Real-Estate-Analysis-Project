```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dfa = pd.read_csv('臺北市.csv')
dfa["type"]= "presale"
dfa
```

## Cleaning presale data in Taipei

```python
def convert_roc_to_temp_gregorian(date_str):
    parts = date_str.split('-')
    year, month, day = parts
    adjusted_year = int(year) + 1911  # Convert to Gregorian year temporarily
    if month  == "00": month = "01"  # Correct invalid month
    if day  == "00": day = "01"  # Correct invalid month
    
    # Validate day and adjust if necessary
    max_days_in_month = 31  # Default to the maximum number of days
    if month in ['04', '06', '09', '11']:  # April, June, September, November
        max_days_in_month = 30
    elif month == '02':  # February
        max_days_in_month = 29 if int(year) % 4 == 0 else 28  # Leap year calculation
    
    day = min(int(day), max_days_in_month)  # Ensure day is within the valid range
    
    return f"{adjusted_year}-{month}-{day:02d}"
dftp = pd.DataFrame(dfa)
dftp["City"]= "Taipei"
dftp.dropna(subset = ["單價","座落"],inplace= True)

dftp[["dist","address","addresss"]] = dftp["座落"].str.split("區", expand=True)
mskk = dftp["addresss"].notna()
dftp.loc[mskk, "address"] = dftp.loc[mskk, "addresss"]


## Information
dftp["p_name"] = dftp["建案名稱"].str.replace("資","").str.replace(" ","")
dftp["tArea"]=dftp["總面積"].str.replace(",","").astype(float)
dftp["price"]=dftp["總價"].str.replace(",","").astype(float)
dftp["h_age"]= 0
dftp["p_price"]=dftp["單價"].astype(int)
dftp["p_lot"] = dftp["車位總價"].notna().astype(int)
dftp["c_price"]=dftp["車位總價"].str.replace(",","").astype(float)
dftp["func"]=dftp["主要用途"]

## Layout
dftp[["temp","bath"]] = dftp["建物格局"].str.replace("衛","").str.split("廳", expand=True).fillna(0)
dftp[["room","temp2"]] = dftp["temp"].str.split("房", expand=True).fillna(0)
dftp[["room","bath"]] =dftp[["room","bath"]].astype(int)


## Level
dftp["樓別/棟高"].replace(regex=r'[\u4e00-\u9fff]+', value='', inplace=True)
dftp[["Fl","tFl"]] = dftp["樓別/棟高"].str.split("/", expand=True).fillna(0)
dftp["Fl"] = pd.to_numeric(dftp["Fl"], errors="coerce").fillna(0).astype(int)
dftp["tFl"] = pd.to_numeric(dftp["tFl"], errors="coerce").fillna(0).astype(int)





dftp["交易日期"]=dftp["交易日期"].str.replace("/","-").apply(convert_roc_to_temp_gregorian)
dftp["date"]=pd.to_datetime(dftp["交易日期"], format='%Y-%m-%d')
dftp["Year"]=dftp["date"].dt.year
dftp.drop(columns =["解約情形","交易明細","功能","備註","index","型態","交易標的","棟及號",
                    "座落","addresss","總面積","單價","交易日期","總價","主建物占比","建物格局",
                   "temp","temp2","建案名稱","樓別/棟高","車位總價","主要建材","主要用途"],inplace= True)
dftp.head()

```

```python
dftp.to_csv("北市預售屋.csv")
```
